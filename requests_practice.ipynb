```python
ages = []

for friend in friends['response']['items']:
    try:
        if len(friend['bdate']) > 6:
            bdate = int(friend['bdate'][-1:-5:-1][::-1])
            age = 2021 - bdate
            ages.append(age)
        else:
            continue
    except:
        pass
```

```python
from collections import Counter


#print(Counter(ages).keys(), Counter(ages).values())

unique_ages = list(set(ages))
answer = []

for age in unique_ages:
    answer.append((age, ages.count(age)))
    
print(answer)
```

```python
answer = sorted(answer, key=lambda x: x[0])
answer = sorted(answer, key=lambda x: x[1], reverse=True)
answer
```
